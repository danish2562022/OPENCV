```python
import numpy as np
import cv2

cap=cv2.VideoCapture(0)
while(True):
    ret,frame=cap.read()
    print(frame)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) == ord('q'):
        print(cv2.waitKey(1))
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
cap.set(3,800)
cap.set(4,900)
while(True):
    ret,frame=cap.read()
    #print(frame)
    frame = cv2.flip(frame,1)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) == ord('q'):
        print(cv2.waitKey(1))
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#Saving a video
cap=cv2.VideoCapture(0)

fourcc=cv2.VideoWriter_fourcc(*'XVID')
out=cv2.VideoWriter('First_Video.avi',fourcc,30.0,(700,500))
while(cap.isOpened()):
    ret,frame=cap.read()
    if ret==True:
        frame=cv2.flip(frame,1)
        out.write(frame)
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) == ord('q'):
        
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
        
        zdfsdasdasdad
```
