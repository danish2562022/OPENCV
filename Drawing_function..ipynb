```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img=np.zeros((512,512,3)) # black image of dimension 512*512 
plt.imshow(img)
```

```python
img_line=cv2.line(img,(0,0),(511,511),(255,0,0),10) # Drawing line from top left corner to bottom right corner
plt.imshow(img_line)
```

```python
# Drawing rectangle
img=np.zeros((512,512,3))
img_rec=cv2.rectangle(img,(384,0),(510,128),(255,0,0),5)
img_rec=cv2.rectangle(img,(384,200),(510,128),(0,255,0),5)
img_rec=cv2.rectangle(img,(175,175),(325,325),(0,0,255),5)
plt.imshow(img)
```

```python
# Drawing circle
img=np.zeros((512,512,3))
img_cir=cv2.circle(img,(250,250),125,(255,0,0),5)
plt.imshow(img_cir)
```
